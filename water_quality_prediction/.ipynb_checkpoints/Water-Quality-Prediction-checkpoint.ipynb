```python
#Importing necessary libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset 

water_data = pd.read_csv (".//water_potability.csv")
```

```python
water_data.info()
```

```python
water_data.head()
```

```python
water_data.shape
```

```python
#EDA Process 

# Check for missing values 

water_data.isna().sum()
```

```python
# Looks like there are a lot number of null values .Removing doesn't look like an option as the dataset will shrink 

water_data.describe()
```

```python
# Imputing with mean value 

water_data.fillna(water_data.mean() , inplace=True)
```

```python
water_data.head()
```

```python
# Lets check the correlation 
sns.heatmap(water_data.corr() , annot = True)
figure = plt.gcf()
figure.set_size_inches( 10 ,6 )
plt.show()

# The variables don't look correlated (or it looks the least)
```

```python
# Let's check if the data is balanced

water_data['Potability'] . value_counts()

```

```python
# Visualising the distribution of data 

water_data.hist(figsize=(15 , 12))
plt.show()

#Data looks almost normally distributed 
```

```python
# Train Test Split 
Y= water_data['Potability']
X = water_data.drop('Potability' , axis=1)
```

```python
X 
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split 

X_train , X_test , Y_train , Y_test = train_test_split (X , Y , random_state = 100 , test_size = 0.2 , shuffle = True)
```

```python
# Decision Tree Classifier 

from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score , confusion_matrix

dt = DecisionTreeClassifier()

dt.fit(X_train , Y_train )

dt_predict = dt.predict (X_test)

accuracy_score (Y_test , dt_predict ) 

confusion_matrix (Y_test , dt_predict )

#Doesn't seem like a great model . Needs Optimisation
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
dt = DecisionTreeClassifier()

criterion = ["gini", "entropy"]
    
splitter = ["best", "random"]

min_samples_split = range(1,12)

gd = dict (criterion = criterion , splitter= splitter , min_samples_split= min_samples_split ) 

rsf = RepeatedStratifiedKFold(n_splits = 3 , random_state =100)


    
```

```python
grid_search = GridSearchCV(estimator = dt , param_grid = gd , scoring ="accuracy" , cv = rsf )
```

```python
grid_search.fit(X_train , Y_train )
```

```python
grid_search.best_params_
```

```python
dt = DecisionTreeClassifier (criterion = 'entropy' , min_samples_split = 11 , splitter = 'best' )
```

```python
dt.fit (X_train , Y_train)
```

```python
dt_predict = dt . predict (X_test)
```

```python
accuracy_score (Y_test , dt_predict ) 
```
